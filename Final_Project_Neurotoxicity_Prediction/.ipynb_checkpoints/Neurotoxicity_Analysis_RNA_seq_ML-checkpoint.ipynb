Below the Dataset 7 and Dataset 8 are downloaded. The information for the chemicals that were used to treat the 3D Neural constructs are also downloaded. Dataset 7 represents Normalized gene expression (TPM) for day 16 neural constructs after dosing with toxic or non-toxic chemicals (two days of exposure). Dataset 8 represents Normalized gene expression (TPM) for day 21 neural constructs after dosing with toxic or non-toxic chemicals (seven days of exposure). 3D Neural constructs were grown for 14 days before chemical exposure.

```python
day_16_df = pd.read_csv('/Users/danielaquijano/Documents/GitHub/Machine-Learning-Course-Projects/Final_Project_Neurotoxicity_Prediction/Final_Project_Files/Day_16_RNA_seq.csv')
day_16_df
```

```python
#Filter dataframes so that only rows of expresison values remain
day_16_filtered_df=day_16_df.iloc[0:19084,2:142]
day_21_filtered_df=day_21_df.iloc[0:19084,2:142]
```

```python
day_16_filtered_df
```

```python
day_21_filtered_df
```

### Obtain Average of Day 16 and Day 21 gene expression

```python
avg_genes=day_21_filtered_df #Fill dataframe with day 21 values as place holders 
```

```python
#Add the corresponding columns from day 16 and day 21 and divide by 2 to obtain average gene expression
for i in range(len(day_21_filtered_df.columns)):
    avg_genes.iloc[:,i] = (day_21_filtered_df.iloc[:,i]+day_16_filtered_df.iloc[:,i])/2
```

```python
pd.set_option('display.max_columns', None) #Shows all the columns 
avg_genes.head() 
```

### Format new Dataframe

```python
#Next step is to edit the column names to leave only the sample (remove day)
#Make a list containing column names
colnameslist=list(avg_genes.columns)
```

```python
#Iterate through list and remove 'day 21' from column name
```

```python
colnames_trim = [re.sub(r'\bday 21 \b', '', item) for item in colnameslist]
```

```python
colnames_trim 
```

```python
avg_genes.columns=colnames_trim 
```

```python
avg_genes
```

```python
#Make two new dataframes for toxic and non toxic exposed samples (for visualization purposes)
avg_genes_nontoxic=avg_genes.iloc[:,0:72]
avg_genes_toxic=avg_genes.iloc[:,72:140]
```

### Re-format data for visualization

```python
#Melt dataframes for visualization
melted_nontoxic=pd.melt(avg_genes_nontoxic)
melted_toxic=pd.melt(avg_genes_toxic)
```

```python
melted_avg_genes=pd.melt(avg_genes)
```

```python
melted_avg_genes
```

### Visualize Data

```python
#Histogram to visualize distribution of gene counts
plt.figure(figsize=(20,5))
plt.xlim(0,400)
count_histogram=sns.histplot(data=melted_avg_genes, x="value",bins=2000,color='pink').set(title='Frequency Distribution of gene counts')
```

```python
#Visualize gene counts for each sample type (toxic/non-toxic compound exposure)
#Visualize non-toxic samples
plt.figure(figsize=(60,20))
plt.ylim(0,125)
boxplot_nontoxic=sns.boxplot(
    data=melted_nontoxic,
    x='variable',
    y='value').set(title='Gene Counts by sample (Non-Toxic Chemical Exposure)')
```

```python
#Visualize toxic samples
plt.figure(figsize=(60,20))
plt.ylim(0,125)

boxplot_toxic=sns.boxplot(
    data=melted_toxic,
    x='variable',
    y='value').set(title='Gene Counts by sample (Toxic Chemical Exposure)')
```

```python
avg_genes_nontoxic.to_csv('avg_genes_nontoxic.csv')
avg_genes_toxic.to_csv('avg_genes_toxic.csv')
avg_genes.to_csv('avg_genes.csv')
melted_nontoxic.to_csv('melted_nontoxic.csv')
melted_toxic.to_csv('melted_toxic.csv')
```
